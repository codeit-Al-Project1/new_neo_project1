```python
import json
import os

def create_category_mapping(json_dir):
    category_mapping = {0: 'background'}  # 0번을 배경으로 미리 지정
    existing_category_ids = set([0])  # 기존 카테고리 ID 확인용 세트
    category_counter = 1

    for root, dirs, files in os.walk(json_dir):
        for file_name in files:
            if file_name.endswith('.json'):
                file_path = os.path.join(root, file_name)

                with open(file_path, 'r', encoding='utf-8') as f:
                    data = json.load(f)

                for category in data.get('categories', []):
                    category_id = category['id']
                    category_name = category['name']

                    # 새 카테고리 ID만 추가
                    if category_id not in existing_category_ids:
                        category_mapping[category_counter] = category_name
                        existing_category_ids.add(category_id)
                        category_counter += 1

    # 마지막에 no_class 추가
    category_mapping[category_counter] = 'no_class'

    return category_mapping

# JSON 파일들이 있는 최상위 디렉토리 경로
json_dir = "C:\\Users\\nihao\\Desktop\\PJ1mine\\data\\train\\train_annotations"

category_mapping = create_category_mapping(json_dir)

# 결과 출력
print("Category Mapping (0: background, 마지막: no_class):")
for idx, name in category_mapping.items():
    print(f"ID {idx}: {name}")
```

```python
import json
import os

def create_category_mapping(json_dir):
    category_mapping = {0: 'background'}
    existing_category_ids = set()
    category_counter = 1

    for root, dirs, files in os.walk(json_dir):
        for file_name in files:
            if file_name.endswith('.json'):
                file_path = os.path.join(root, file_name)

                with open(file_path, 'r', encoding='utf-8') as f:
                    data = json.load(f)

                for category in data.get('categories', []):
                    category_id = category['id']
                    category_name = category['name']

                    # 이름 앞에 번호 포함하여 중복 체크
                    formatted_name = f"{category_id} {category_name}"

                    if formatted_name not in existing_category_ids:
                        category_mapping[category_counter] = formatted_name
                        existing_category_ids.add(formatted_name)
                        category_counter += 1

    # 마지막에 no_class 추가
    category_mapping[category_counter] = 'no_class'

    return category_mapping

# JSON 파일들이 있는 최상위 디렉토리 경로
json_dir = "C:\\Users\\nihao\\Desktop\\PJ1mine\\data\\train\\train_annotations"

category_mapping = create_category_mapping(json_dir)

# 결과 출력
print("Category Mapping (0: background, 마지막: no_class):")
for idx, name in category_mapping.items():
    print(f"ID {idx}: {name}")

# 정확한 개수 확인
print(f"\n총 클래스 개수: {len(category_mapping)}개")
```
